 # 컴퓨터 정수와 이진수

 컴퓨터 정수의 표현은 int로 하며 이진수로 표현되게 된다.
 
 이때 컴퓨터의 word 길이 32bit니 64bit니 하는 것이
 
 컴퓨터 정수값의크기(주소값) 이 된다. 
 
 00000000~ 11111111 까지 1byte를 가짐
 
 이를 16진수 두 자리로 표현하기도 함
 
 32 bit를 기준으로하면 정수는 4byte를 사용하게 되고
 
 이 때 메모리의 가장 작은단위가 1byte로 정수는 4칸을 차지하게 되는데 보통
 
 16진수를 사용하여 0A1B2C30와 같은 식으로 8자리로 표현하게 된다.
 
 이 때 앞의 단위인 '0A' 같은 큰 단윗수를 가지는 부분을 
 
 MSB (Most Significant Byte)라 부르고 최상위 바이트라고 하고
 
 '30' 같이 가장 작은 자릿수를 담당하는 부분을
 
 LSB (Least Significant Byte) 최하위 바이트라고 부른다.
 
 이떄 MSB 부분부터 먼저 읽는 방식을 빅 엔디안이라고 하고
 
 LSB 부분부터 먼저 읽는 방식을 리틀 엔디안이라고 부른다.
 

이 때 빅엔디안은 사람이 읽는 것처럼 자릿수가 높은것부터 읽기 때문에 디버깅 하거나 할때 직관적이며

알아보기 편한 장점이 있으며 

리틀 엔디안의 경우 하위 값만 얻으면 되는 경우 별도의 계산이 필요없어 하위 1바이트, 2바이트만 사용하여

값을 얻을 수 있지만 빅 엔디안의 경우 뒤에 앞에 00을 붙여줘야 하위 값을 얻을 수 있기 때문에 단점이 있다고 한다.

네트워크에서는 빅 엔디안을 쓰도록 규약이 되어있어 리틀엔디안의 경우 빅엔디안으로 변환하여야 한다고 한다.
